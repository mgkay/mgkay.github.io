```julia
function isorigin(rte::Vector{Int})
    # Input error checking (assuming checkrte is defined elsewhere)
    checkrte(rte, nothing, true)

    seen = Set{Int}()
    is = [!in(push!(seen, rte[i])) for i in 1:length(rte)]  # Inline boolean assignment
    
    return is
end
```

```julia
function rte2idx(rtein::Union{Vector{Int}, Vector{Vector{Int}}})
    # Input error checking, assuming checkrte is defined elsewhere
    checkrte(rtein)

    # Ensure that rtein is always treated as a collection of routes
    rte = if !isa(rtein, Vector{Vector})
        [rtein]  # Wrap the single route in a vector if it's not a vector of vectors
    else
        rtein
    end

    # Initialize idx as a vector of vectors, matching the size of rte
    idx = Vector{Vector{Int}}(undef, length(rte))
    
    # Iterate over each route and find indices where isorigin is true
    for i in 1:length(rte)
        idx[i] = rte[i][isorigin(rte[i])]
    end

    # If the input was not a vector of vectors, return the first (and only) element
    return isa(rtein, Vector{Vector}) ? idx : idx[1]
end
```

```julia
function rte2loc(rtein::Union{Vector{Int}, Vector{Vector{Int}}}, sh::Dict, tr::Union{Nothing, Dict} = nothing)
    checkrte(rtein, sh)  # Input error checking (assumed defined elsewhere)

    rte = isa(rtein, Vector{Vector}) ? rtein : [rtein]  # Ensure rte is a vector of vectors
    loc = Vector{Union{Vector{Int}, Nothing}}(undef, length(rte))  # Preallocate location vector

    for (i, r) in enumerate(rte)
        if any(isnan.(r))
            loc[i] = nothing
            continue
        end

        # Determine origin locations and destination locations
        is = isorigin(r)
        loc[i] = getindex.(sh[r[is]], "b") # Origin locations
        append!(loc[i], getindex.(sh[r[.!is]], "e")) # Destination locations

        # Handle optional truck locations
        if tr !== nothing
            bloc = get(tr, "b", loc[i][1])
            eloc = get(tr, "e", loc[i][end])
            if bloc == eloc && sum(diff(bloc .== loc[i]) .!= 0) > 2
                loc[i] = nothing
            else
                loc[i] = [bloc; loc[i]; eloc]
            end
        end

        loc[i] = loc[i]'  # Ensure it's a row vector
    end

    return isa(rtein, Vector{Vector}) ? loc : loc[1]  # Unwrap single route vector if needed
end
```

```julia
function loccost(loc::Vector{Int}, C::AbstractMatrix)
    n = length(loc) - 1
    c = Vector{eltype(C)}(undef, n)
    for i in 1:n
        c[i] = C[loc[i], loc[i+1]]
    end
    return c
end

loc = [1, 2, 4, 3]
C = [0 2 5 13;
     3 0 10 8;
     4 15 0 12;
     7 6 9 0]
c = loccost(loc, C)
```

```julia
function twoopt(tour::Vector{Int}, dist_matrix::Matrix{Float64})
    # Number of cities (excluding the return to the starting city)
    n = length(tour) - 1  # Exclude the repeated city at the end
    
    best_tour = copy(tour)   # Current best tour
    best_distance = tour_distance(best_tour, dist_matrix)  # Distance of the best tour
    
    # Loop over all pairs of indices (i, k) such that 1 <= i < k <= n
    improvement = true
    while improvement
        improvement = false
        for i in 2:(n-2)
            for k in (i+2):(n)
                # Apply two-opt swap between i and k
                new_tour = twoopt_swap(best_tour, i, k)
                println(new_tour)
                new_distance = tour_distance(new_tour, dist_matrix)
                
                # If the new tour is shorter, accept the new tour
                if new_distance < best_distance
                    best_tour = new_tour
                    best_distance = new_distance
                    improvement = true
                end
            end
        end
    end
    
    return best_tour, best_distance
end

# Function to calculate the total distance of a given tour, accounting for asymmetric distances
function tour_distance(tour::Vector{Int}, dist_matrix::Matrix{Float64})
    total_distance = 0.0
    for i in 1:(length(tour)-1)
        total_distance += dist_matrix[tour[i], tour[i+1]]
    end
    return total_distance
end

# Function to perform the two-opt swap
function twoopt_swap(tour::Vector{Int}, i::Int, k::Int)
    new_tour = vcat(tour[1:i-1], reverse(tour[i:k]), tour[k+1:end])
    return new_tour
end
```

```julia
# Define a sample tour
tour = [1, 2, 4, 6, 3, 5, 1]  # Note the return to the start

# Example distance matrix
dist_matrix = [
    0.0  2.0  9.0  10.0  7.0  3.0;
    2.0  0.0  5.0   6.0  3.0  4.0;
    9.0  5.0  0.0   8.0  7.0  6.0;
    10.0 6.0  8.0   0.0  4.0  5.0;
    7.0  3.0  7.0   4.0  0.0  2.0;
    3.0  4.0  6.0   5.0  2.0  0.0
]

# Apply two-opt improvement
best_tour, best_distance = twoopt(tour, dist_matrix)

println("Improved tour: ", best_tour)
println("Total distance: ", best_distance)
```

```julia
function twoopt(r, rTCh)
    rᵒ, TCᵒ = copy(r), rTCh(r)
    println("*", TCᵒ, ": ", rᵒ)
    done = false
    while !done
        done = true
        for i = 1:length(r)-2
            for j = i+2:min(length(r)-1, length(r)+i-3)
                r = vcat(r[1:i], reverse(r[i+1:j]), r[j+1:end])
                TC = rTCh(r)
                if TC < TCᵒ
                    rᵒ, TCᵒ = r, TC
                    println("*", TCᵒ, ": ", rᵒ)
                    done = false
                    break
                else
                    println(" ", TC, ": ", r)
                end
            end
            if done == false
                break
            end
        end
    end
    return rᵒ, TCᵒ
end
#twoopt([1, 2, 4, 6, 3, 5, 1], rTCh)    
    
```

```julia
function dgc(xy₁, xy₂; unit=:mi)
    length(xy₁) == length(xy₂) == 2 || error("Inputs must have length 2.")
    unit in [:mi, :km] || error("Unit must be :mi or :km")

    Δx, Δy = xy₂[1] - xy₁[1], xy₂[2] - xy₁[2]
    a = sind(Δy / 2)^2 + cosd(xy₁[2]) * cosd(xy₂[2]) * sind(Δx / 2)^2
    2 * asin(min(sqrt(a), 1.0)) * (unit == :mi ? 3958.75 : 6371.00)
end

Dgc(X₁, X₂) = [dgc(i, j) for i in eachrow(X₁), j in eachrow(X₂)]

XY = hcat(df.LON, df.LAT)
D = Dgc(XY, XY)
```

```julia
segcost(r, C) = map(i -> C[r[i], r[i+1]], 1:length(r) - 1)

rTCh(r) = sum(segcost(r, D))

function twoopt(r, rTCh)
    rᵒ, TCᵒ = copy(r), rTCh(r)
    done = false
    while !done
        done = true
        for i = 1:length(r)-2
            for j = i+2:min(length(r)-1, length(r)+i-3)
                r′ = vcat(r[1:i], reverse(r[i+1:j]), r[j+1:end])
                TC = rTCh(r′)
                if TC < TCᵒ
                    rᵒ, TCᵒ = r′, TC
                    done = false
                    break
                end
            end
            if done == false
                break
            end
        end
        r = copy(rᵒ)
    end
    return rᵒ, TCᵒ
end
```

```julia
include("rtefun.jl")

function rte2ldmx(rte)
    is = isorigin(rte)
    if count(!iszero, diff(is)) == 1   # Non-interleaved route => one load mix
        return [unique(rte)]
    else                               # Interleaved route
        ldmx = [Int[]]
        for i in 1:length(rte) - 1
            if is[i]
                push!(ldmx[end], rte[i])
                if !is[i + 1]
                    push!(ldmx, copy(ldmx[end]))
                end
            else
                filter!(!=(rte[i]), ldmx[end])
            end
            println(ldmx)
        end
        return ldmx[1:end-1]
    end
end

#ldmx = rte2ldmx([2, 1, 3, 2, 1, 4, 3, 4])
ldmx = rte2ldmx([1:5; 1:5;])
#rte2ldmx([1, 2, 1, 3, 3, 2])
```
