### Ex: Truck Shipment Example

* Product shipped in cartons from Raleigh, NC (27606) to Gainesville, FL (32606)
* Each identical carton weighs 40 lb and occupies 9 ft<sup>3</sup> (its cube)
* Donâ€™t know the linear dimensions of each unit for TL and LTL
* Cartons can be stacked on top of each other in a trailer
* Additional info/data is presented only when it is needed to determine the answer

__1.__ Assuming that the product is to be shipped P2P TL, what is the maximum payload for each trailer used for the shipment?



```julia
uwt = 40                                                     # lb
ucu = 9                                                      # ft3
s = uwt/ucu                                                  # lb/ft3
Kwt = 25                                                     # ton
Kcu = 2750                                                   # ft3
qmax = min(Kwt, s*Kcu/2000)                                  # ton
println("Density = ", s, "\nMax payload = ", qmax)
```

__2.__ On Jan 10, 2018, 300 cartons of the product were shipped. How many truckloads were required for this shipment?

```julia
udTL = 300                                                   # cartons
qTL = udTL*(uwt/2000)                                        # ton
println("Shipment size = ", qTL, "\nNo. TLs = ", ceil(Int, qTL/qmax))
```

__3.__ Before contacting the carrier to negotiate (and using Jan 2018 PPI), what would have been the estimated TL transport charge for this shipment?



```julia
d = 532                                                      # Google Maps road distance
ppiTL = 131.0                                                # TL PPI for Jan 2018
rTL = 2.00 * ppiTL/102.7                                     # $/mi
cTL = ceil(qTL/qmax) * rTL * d                               # $
println("rTL = ", rTL, "\ncTL = ", cTL)
```

__4.__ Using the Jan 2018 PPI LTL rate estimate, what was the transport charge to ship 15 cartons LTL?



```julia
udLTL = 15                                                   # cartons
qLTL = udLTL*(uwt/2000)                                      # ton
ppiLTL = 177.4                                               # LTL PPI for Jan 2018

rateLTL(q,s,d,ppi) = ppi*(s^2/8 + 14)/((q^(1/7) * d^(15/29) - 7/2) * (s^2 + 2*s + 14))

rLTL = rateLTL(qLTL,s,d,ppiLTL)                              # $/ton-mi
cLTL = rLTL * qLTL * d                                       # $
println("qLTL = ", qLTL, "\nrLTL = ", rLTL, "\ncLTL = ", cLTL)
```

__5.__ What would the shipment size have to be so that the TL and LTL charges are equal?



```julia
using Optim

cTLh(q) = ceil(q/qmax) * rTL * d
cLTLh(q) = rateLTL(q,s,d,ppiLTL) * q * d
qI = optimize(q -> abs(cTLh(q) - cLTLh(q)), 0, qmax).minimizer
println("qI = ", qI)
```

```julia
# Using LB = 0, UB = qmax for 1-D bounded interval search instead of Nelder-Mead
optimize(q -> abs(cTLh(q) - cLTLh(q)), 0, qmax)
```

_Brent's method_ (1973) is the default method used for univariate optimization in many software packages including `Optim` (Nelder-Mead is the default for multivariate optimization). It uses parabolic interpolation whenever possible because it can converge to the optimum solution much faster than interval search.

#### Univariate Optimization

$ \begin{eqnarray}
x^* &=& \mathrm{arg}\underset{x}{\operatorname{min}} \bigl\{ \, f(x) : L\!B \leq x \leq U\!B \bigr\} \\
T\!C^* &=& f(x^*)
\end{eqnarray}$



__6.__ What are the TL and LTL minimum charges?

 

```julia
MC_TL = 45*(rTL/2)
MC_LTL = (ppiLTL/104.2) * (45 + d^(28/19)/1625)
println("MC_TL = ", MC_TL, "\nMC_LTL = ", MC_LTL)
```

#### Independent Transport Charge

```julia
using CairoMakie
dcf() = display(current_figure())

c0h(q) = min(max(cTLh(q), MC_TL), max(cLTLh(q), MC_LTL))

q = 1/2000:0.01:8                                 # Plot charge from 1 lb to 8 tons
h = lines(q, [c0h(i) for i in q])
h.axis.xlabel = "Shipment Size (ton)"
h.axis.ylabel = "Transport Charge (\$)"
h.axis.title = "Independent shipment charge: Class 200 from 27606 to 32606"
dcf();
```

__8.__ Using the same LTL shipment, what is the transport cost found using the undiscounted CzarLite tariff?




```julia
q = qLTL
i = 2
disc = 0
MC = 95.23
ODi = 127.69
ci = ODi*20*q
ODip1 = 99.92
qBi = 0.5
cip1 = ODip1*20*qBi
c_tar = (1 - disc)*max(MC, min(ci, cip1))
println("ci = ", ci, "\ncip1 = ", cip1, "\nc_tar = ", c_tar)
```
