__(1)__ A new tool crib is to be located on the floor of a facility. Workers from eight different workstations will make 11, 46, 36, 20, 32, 23, 20, and 18 trips per shift to the tool crib. The 2-D locations of the workstations are shown in the figure below. Assuming rectilinear distance is a reasonable approximation of the actual travel distance, determine the optimal location for the tool crib that will minimize the total distance that the workers travel.




```julia
using Optim

pt = [(85,40), (75,40), (75,85), (45,35), (40,55), (35,55), (30,85), (55,90)]
w = [11, 46, 36, 20, 32, 23, 20, 18]
d1(x1, x2) = length(x1) == length(x2) ? sum(abs.(x1 .- x2)) : 
    error("Inputs not same length.")
TC(xy) = sum(w .* d1.([xy], pt))
xyᵒ = optimize(TC, [0., 0.]).minimizer
```

__(2)__ A product will be produced at a single plant. Each ton of the product requires 2 tons of raw material A from a supplier located in Durham on the figure below, 0.55 tons of raw material B from a supplier located in Statesville, 1.2 tons of raw material C from a supplier located in Wilmington, and 0.35 tons of raw material D from a supplier located in Asheville. The annual demand for the product for the four customers in Asheville, Winston-Salem, Greensboro, and Raleigh are 10, 38, 20, and 46 tons, respectively. The cost per ton-mile to ship the raw materials A, B, C, and D to the plant is \\$0.08, \\$0.05, \\$0.15, and \\$0.03, respectively, and the cost per ton-mile to ship the finished goods from the plant to the customers is \\$0.80. All scrap is disposed of locally. (a) Assuming that I-40 will be used for all travel, where should the plant be located to minimize transportation costs? (b) With respect to its monetary weight, is the product weight gaining or weight losing?



```julia
# (a) First, determine monetary weights
# Outbound
fout = [10, 38, 20, 46]        # ton/yr
rout = 0.8                     # $/ton-mi
@show wout = fout * rout       # $/mi-yr
# Inbound
BOM = [2, 0.35, 1.2, 0.55]     # Bill-of-material ratio
@show fin = BOM * sum(fout)    # ton/yr
rin = [0.08, 0.05, 0.15, 0.03] # $/ton-mi
@show win = fin .* rin         # $/mi-yr
w = vcat(win, wout)            # $/mi-yr
```

```julia
# (a) Next, determine location
pt = [(270,), (150,), (420,), (50,), (50,), (190,), (220,), (295,)]
d1(x1, x2) = length(x1) == length(x2) ? sum(abs.(x1 .- x2)) : 
    error("Inputs not same length.")
TC(xy) = sum(w .* d1.([xy], pt))
xyᵒ = optimize(TC, [0.]).minimizer
```

```julia
# (b) Weight gaining or weight losing
println("The product is ", sum(win) > sum(wout) ? "weight losing." : "weight gaining.")
```

__(3)__ Create ten random (real-number) points in the region defined by the rectangle `[(2, 1), (10, 7)]` using the random number seed `2073`.

```julia
using Random
Random.seed!(2073)

r = [(2, 1), (10, 7)] 
(xmn, ymn), (xmx, ymx) = r
pt = [(xmn + rand()*(xmx - xmn), ymn + rand()*(ymx - ymn)) for _ in 1:10]
```

```julia
using CairoMakie

unzip(x) = collect.(zip(x...))
dcf() = display(current_figure())

scatter(unzip(pt)...)
dcf();
```

(a)	Determine the location that minimizes the sum of the squared-Euclidean distances.

```julia
# First approach: Calculate the centroid
using Statistics

xᵒ = mean.(unzip(pt))
```

```julia
# Second approach: Minimize the sum of squared-euclidean distances
using Optim

d2(x1, x2) = length(x1) == length(x2) ? sqrt(sum((x1 .- x2).^2)) : 
    error("Inputs not same length.")

x₀ = [0., 0.]
xᵒ = optimize(x -> sum(d2.([x], pt).^2), x₀).minimizer
scatter!(xᵒ...; marker=:star5, color=:green, label="Centroid")
xᵒ
```

(b)	Determine the location that minimizes the sum of the Euclidean distances.

```julia
# Minisum
x₀ = mean.(unzip(pt))
xᵒ = optimize(x -> sum(d2.([x], pt)), x₀).minimizer
scatter!(xᵒ...; marker=:star5, color=:red, label="Minisum")
xᵒ
```

(c)	Determine the location that minimizes the maximum Euclidean distance.

```julia
# Minimax
xᵒ = optimize(x -> maximum(d2.([x], pt)), x₀).minimizer
scatter!(xᵒ...; marker=:star5, color=:cyan, label="Minimax")
axislegend()
dcf()
xᵒ
```

(d)	Determine the location that minimizes the sum of the Euclidean distances, with the restriction that the location cannot be within the rectangle defined by `[(5,3) and (7,5)]`. The straight line from the location to any point can cross the rectangle (e.g., if the location restriction is due to zoning, then travel across the zone can still occur).

```julia
r = [(5, 3), (7, 5)]   # SW and NE corners of feasible region
isinrect(x, r) = r[1][1] <= x[1] <= r[2][1] && r[1][2] <= x[2] <= r[2][2]
TC(x) = isinrect(x, r) ? Inf : sum(d2.([x], pt))
x₀ = [6., 4.]
res = optimize(TC, x₀)
@show xᵒ = res.minimizer
res
```

```julia
TC(x₀)   # x₀ inside restricted region, no progress in deducing TC results in failure 
```

```julia
if Optim.converged(res) == false      # Can test for failure in code
    println("Optimization failed.")
end
```

```julia
x₀ = [0., 0.]            # x₀ outside restricted region
res = optimize(TC, x₀)
@show res.minimizer
res
```

(e)	Determine the location that maximizes the minimum Euclidean distance, with the restriction that the location must be within the rectangle defined by `[(2,1) and (10,7)]`.

```julia
r = [(2, 1), (10, 7)]   # SW and NE corners of feasible region
isinrect(x, r) = r[1][1] <= x[1] <= r[2][1] && r[1][2] <= x[2] <= r[2][2]
TC(x) = isinrect(x, r) ? -minimum(d2.([x], pt)) : Inf

xrng = r[1][1]:0.1:r[2][1]
yrng = r[1][2]:0.1:r[2][2]
Z = [TC((x, y)) for x in xrng, y in yrng]
contour(xrng, yrng, Z; levels = 10)
scatter!(unzip(pt)...)
dcf();
```

```julia
# Approach: Use each EF as x₀ and keep the best
TCᵒ, xᵒ = Inf, nothing
for x₀ in pt
    res = optimize(TC, collect(x₀))
    if res.minimum < TCᵒ
        TCᵒ, xᵒ = res.minimum, res.minimizer
        println("Local optimum: ", xᵒ)
    end
end
scatter!(xᵒ...; marker=:star5, color=:red)
dcf()
xᵒ
```
