Assuming that transport costs are proportional to great circle distance and that each DC can supply any amount of the product, determine the amount of product that should be supplied from each DC to each customer in order to minimize total ton-miles.

```julia
using DataFrames, Logjam.DataTools

function zcta5_2lonlat(zcta5)
    df = uszcta5()
    idx = findfirst(r -> r.ZCTA5 == zcta5, eachrow(df))
    if idx === nothing
        error("'$zcta5' not found in DataFrame")
    end
    return collect(df[idx, [:LON, :LAT]])
end

XYS = vcat((zcta5_2lonlat(i)' for i in [30669, 32606, 30733, 23830, 23061])...)
XYC = vcat((zcta5_2lonlat(i)' for i in [20548, 26149, 36317])...)
```
